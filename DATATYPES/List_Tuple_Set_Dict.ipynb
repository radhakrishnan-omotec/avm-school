# DataType


```python
Data Types are the classification or categorization of data items.
```

```python
DataType :
    Numeric 
    Sequence Type
    Boolean
    Set
    Dictionary
```

### Numeric 

```python
Integers – This value is represented by int class.
           It contains positive or negative whole numbers (without fractions or decimals).
           For example – 2
    
Float    – This value is represented by the float class. It is a real number with a floating-point representation.
           It is specified by a decimal point.
           For example – 2.8
    
Complex Numbers – Complex number is represented by a complex class. It is specified as (real part) + (imaginary part)j.
                  For example – 2+3j
```

```python
integer_value=55   # 23 67 8999
print(type(integer_value))
print(integer_value)
```

```python
float_value=5.6
print(type(float_value))
print(float_value)
```

```python
complex_value=3+5j
print(type(complex_value))
print(complex_value)
```

### Sequence Type

#### String

```python
String -  A string is a collection of one or more characters put in a single quote, double-quote, or triple-quote. 
```

```python
var='string is a datatype in python'
print(var)
print(type(var))
```

```python
var1="Python Datatype"
print(var1)
print(type(var1))
```

```python
var2='''Python works on different platforms (Windows, Mac, Linux, Raspberry Pi, etc).
        Python has a simple syntax similar to the English language.'''
print(var2)
print(type(var2))
```

#### List

```python
Lists - List in Python can be created by just placing the sequence inside the square brackets[]
```

```python
Rollno_list =[101,102,103,104,105]
print(Rollno_list)
print(type(Rollno_list))
```

```python
school =["ground","teacher",102,"lab",1.0,2+4j]
print(school)
print(type(school))
```

#### Tuple

```python
Tuple - Tuple in Python can be created by just placing the sequence inside the square brackets()
```

```python
Rollno_tuple =(101,102,103,104,105)
print(Rollno_tuple)
print(type(Rollno_tuple))
```

```python
school =("ground","teacher",102,"lab",1.0,2+4j)
print(school)
print(type(school))
```

### set
